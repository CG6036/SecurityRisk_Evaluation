```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load Data
df0 = pd.read_csv('dataset0.csv')
df1 = pd.read_csv('dataset1.csv')
df2 = pd.read_csv('dataset2.csv')
df3 = pd.read_csv('dataset3.csv')
# Merge Data 
df_merge = pd.concat([df0.iloc[:,1:], df1.iloc[:,1:], df2.iloc[:,1:], df3.iloc[:,1:]], ignore_index = True)
```

```python
from collections import Counter
import re

# Function to count words in the 'full_log' column
def count_words(text):
    # Use regex to find words (alphanumeric characters including underscores)
    words = re.findall(r'\w+', text.lower())  # Convert to lowercase to count all variations of a word the same
    return words
```

### 1) Explore Text data

```python
def preProcess(df):
    df['log_noNum'] = df['full_log'].apply(lambda x: re.sub(r'\d+', '', x))
    df['log_noNum']=df['log_noNum'].apply(lambda x: re.sub(r'(\\n)',' ',x))
    df['log_noNum']=df['log_noNum'].apply(lambda x: re.sub(r'[^a-zA-Zㄱ-ㅣ가-힣0-9:=\s\(\)./,\<\>]+','',x))
    df['log_noNum']=df['log_noNum'].apply(lambda x: re.sub(r' ?(?P<note>[:=\(\)./,\<\>]) ?', ' \g<note> ', x))
    return df
```

```python
df0 = preProcess(df0)
df1 = preProcess(df1)
df2 = preProcess(df2)
df3 = preProcess(df3)
```

```python
# Most Common Words in df0
word_counts0 = Counter()
df0['log_noNum'].apply(lambda x: word_counts0.update(count_words(x)))
top_10_df0 = pd.DataFrame(word_counts0.most_common(10), columns=['Word', 'Frequency'])
# Most Common Words in df1
word_counts1 = Counter()
df1['log_noNum'].apply(lambda x: word_counts1.update(count_words(x)))
top_10_df1 = pd.DataFrame(word_counts1.most_common(10), columns=['Word', 'Frequency'])
# Most Common Words in df2
word_counts2 = Counter()
df2['log_noNum'].apply(lambda x: word_counts2.update(count_words(x)))
top_10_df2 = pd.DataFrame(word_counts2.most_common(10), columns=['Word', 'Frequency'])
# Most Common Words in df3
word_counts3 = Counter()
df3['log_noNum'].apply(lambda x: word_counts3.update(count_words(x)))
top_10_df3 = pd.DataFrame(word_counts3.most_common(10), columns=['Word', 'Frequency'])
```
